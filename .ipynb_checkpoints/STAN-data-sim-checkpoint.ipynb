```python
import pystan
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import CategoricalEncoder
```

```python
# Function to create a set of unique names from the name List
def select_names(num_needed):
    name_file = open("names.txt", "r")
    names = name_file.readlines()
    num_of_names = len(names)
    
    np.random.seed(1997)
    first_indices = np.random.choice(num_of_names, num_needed, replace=False)
    np.random.seed(2002)
    second_indices = np.random.choice(num_of_names, num_needed, replace=False)
    
    names = [(names[a][:-1],names[b][:-1]) for a,b in zip(first_indices, second_indices)]
    return names
```

```python
# generates a random subset of classes that student is taking, generates a distribution for the student pulled from the hyperprior
student_records = []

for student in select_names(100):
    for class_name in classes.keys():
        
        taken = np.random.binomial(1, 0.9, len(classes[class_name].keys())) # selects random subset of classes student has taken
        
        for ind, assignment in enumerate(classes[class_name]):
            if taken[ind] == 1:
                assign = classes[class_name][assignment]
                num_entered = int(np.random.normal(assign["num_avg"], assign["num_sd"], 1))
                num_entered = num_entered if num_entered > 0 else 0 # checks if estimated value is negative, converts it to 0
                
                # generate distribution specific to student
                mu = np.random.normal(assign["mu"], assign["sd"], 1)
                sd = np.random.exponential(assign["sd"], 1)
                
                student_entered_times = np.random.normal(mu, sd, num_entered)
                if len(student_entered_times) != 0:
                    for entry in student_entered_times:
                        record = []
                        record.append(student[0])
                        record.append(student[1])
                        record.append(class_name)
                        record.append(assignment)
                        record.append(abs(entry))
                        student_records.append(record)
                    
                
            
            
```

```python
df = pd.DataFrame(student_records)
```

```python
# add column names to dataframe
df.columns = ['first_name', 'last_name', 'class_type', 'assign_type', 'user_time_est', 'hard_easy', 'confidence', 'hard_easy_expected', 'actual_time']
```

```python
# break categorical variables into dummy variables
df = pd.get_dummies(df, columns=["class_type", "assign_type", "hard_easy", "confidence", "hard_easy_expected"])
```

```python
# dictionary to contain actual population parameters
param_vals = [1.05, -100, 10, -7.5, 25, 15, -12, 30, 0.5, -5, 10, 12, 20, -5, 15, 27, 6, 4.5, -10, -5, 0, 8, 12, 12, 3, 0.05, -4, -12, 15, 6, -0.01, -5, -10]

edu_params = {}
for index, param in enumerate(df.columns[2:]):
    if param != 'actual_time':
        edu_params[param]= param_vals[index]
        print param + ': ' + str(param_vals[index])
```

```python
# create response var
alpha =0.05
sigma = 10
```

```python
response = np.zeros(len(df))
for key in edu_params.keys():
    tmp = [val*edu_params[key] for val in df[key]]
    response += tmp
response = np.absolute(response)
df['actual_time'] = response + alpha
```

```python
# introduce error
df['actual_time'] = np.random.normal(df['actual_time'], sigma)
```

# Pystan!

```python
import pystan
```

```python
df.columns
```

```python
model= """
data {
    int<lower=0> N;
    int<lower=0> K;
    matrix[N,K] X;
    vector[N] y;
}
parameters {
    vector[K] beta;
    real<lower=0> sigma;
}
model {
    y ~ normal(X*beta, sigma);
}
"""
```

```python
df2 = df.drop('first_name', axis=1).drop('last_name', axis=1)
response = df['actual_time']
df2 = df2.drop('actual_time', axis=1)
```

```python
df2_mat = pd.DataFrame.as_matrix(df2)
response = list(response)
```

```python
data = {
    'N': len(df2),
    'K':len(df2.columns),
    'X': df2_mat,
    'y': response
}
```

```python
# Compile the model
sm = pystan.StanModel(model_code=model)
```

```python
# Train the model and generate samples
fit = sm.sampling(data=data, iter=500, chains=2, warmup=100, thin=1, seed=101)
```
